### Remove all low resolution of image

```python
import os
from PIL import Image

# Define the directory containing the images
dir = r".\dataset\Men"

types = ['Heart', 'Round', 'Square', 'oblong', 'Oval']

# Loop through all types
for type in types:
    dir_path = os.path.join(dir, type)
    file_names = os.listdir(dir_path)

    for file_name in file_names:
        # Construct the file path
        file_path = os.path.join(dir_path, file_name)

        # Open the image to check its size
        try:
            with Image.open(file_path) as img:
                width, height = img.size

            # Check if either dimension is below 500 pixels
            if width < 200 or height < 200:
                # Remove the image
                os.remove(file_path)
                print(f"Removed {file_name} due to its size being below 500 pixels in either dimension.")
        except IOError:
            print(f"Error opening or reading {file_name}. Skipping.")

print("Image removal process completed.")
```

### Move to train test folder separately

```python
import os
import shutil
import random

# Base directory containing the images
base_dir = './dataset/Men/'
test_dir = './dataset/Men/test'
train_dir = './dataset/Men/train'
folders = ["Heart", "Oblong", "Oval", "Round", "Square"]

for folder in folders:
    dir_path = os.path.join(base_dir, folder)

    # Ensure train and test directories exist
    train2_dir = os.path.join(train_dir, folder)
    test2_dir = os.path.join(test_dir, folder)
    os.makedirs(train2_dir, exist_ok=True)
    os.makedirs(test2_dir, exist_ok=True)

    # Get a list of all file names in the directory
    file_names = os.listdir(dir_path)
    file_names = [f for f in file_names if os.path.isfile(os.path.join(dir_path, f))]

    # Randomly select 150 images for training and 50 for testing
    train_files = random.sample(file_names, 150)
    remaining_files = set(file_names) - set(train_files)
    test_files = random.sample(list(remaining_files), 50)

    # Move the selected files to their respective folders
    for file_name in train_files:
        shutil.move(os.path.join(dir_path, file_name), os.path.join(train_dir, file_name))

    for file_name in test_files:
        shutil.move(os.path.join(dir_path, file_name), os.path.join(test_dir, file_name))

print("Images have been split into train and test folders.")
```
